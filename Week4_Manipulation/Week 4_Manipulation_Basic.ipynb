## **1.1 Basic concepts of manipulation**

* **Interact with the environment** and **modify** it using a robot to
perform useful tasks

  * pick and place an object.

  * fasten two parts of a car chassis.


* An integral part of automation.

## **1.2 Manipulation - functional modules**

Manipulation consists of multiple successive steps. These steps are also known as functional modules.

1. We need to **sense** the environment somehow.

2. We need to **plan** the motion we wish to perform.

3. We need to **check** we won't collide with anything while moving.

4. We need to **calculate** our trajectory and how fast we want to go.

## **`Question 1`**

* It is always **necessary to have an environment perception module** to be able to successfully manipulate object ? 

* Answer: **False**

* **Reason:** We can also manipulate objects if their locations are pre-programmed and known **beforehand**. 

 * For example, in a car manufacturing plant, some of the car parts always arrive at a precise location and go to a precise location.

## **2.2 MoveIt! - glues modules together**

image.png

Per functional modules, there are many software options : 

1. **Sensors**: 
 *  [Real Sense](http://wiki.ros.org/RealSense) 
 *  The [Ensenso 3D camera](https://github.com/ensenso/ros_driver)

2. **Motion planning**: **`OMPL`** or **`CHOMP`**

3. **Collision Checking**:
 *  [Flexible Collision Library (**`FCL`**)](https://github.com/flexible-collision-library/) 
 * [Bullet Physics Library](https://pybullet.org/Bullet/BulletFull/), . 

4. **Trajectory processing**: 
 * Iterative Parabolic Time Parameterization (**`IPTP`**) 
 * [Time Optimal Path Paramaterization] (https://github.com/quangounet/TOPP)

6. **Motion Execution**: Various robot drivers that are developed and supported by the ROS-Industrial consortium



## **2.3 MoveIt! - A lot goes on under the hood**

1. Maintains information consistency between our ROS application and the different functional modules..

2. Integrate robot **`kinematic information with planning`**.

3.  Report and request **`alternative motion plans`** in case of collisions.

  * Takes care of the interaction between **collision detection** and motion  planning.

4. Account for any hardware limitations such as **`joint limits`**.

5. Keep track of the current state of the robot and its environment
while performing a manipulation task.
 * Keeps track of Robot location.


6. Talk to the robot `hardware/simulation` and 

7. Notify the ROS application once a desired manipulation task is complete.
